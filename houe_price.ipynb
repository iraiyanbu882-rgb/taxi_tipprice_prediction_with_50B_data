```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing  import OneHotEncoder
from sklearn.metrics import r2_score,root_mean_squared_error
from sklearn.pipeline import Pipeline
import pickle as pk
```

```python
df=pd.read_csv('chennai_house_price.csv')
df[:5]
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['age'].mean()
```

```python
df['age']=df['age'].fillna(1)
df['bathroom']=df['bathroom'].fillna(2)
```

```python
df['bathroom'].mean()
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)
area_encoded=encoder.fit_transform(df[['location']])
print(area_encoded[:5])
```

```python
cols=encoder.get_feature_names_out(['location'])
```

```python
encoded_df = pd.DataFrame(area_encoded, columns=cols)
final_df = pd.concat([df.drop('location', axis=1), encoded_df], axis=1)
```

```python
final_df[:5]
```

```python
target=final_df['price(in lakhs)']
feartures=final_df.drop(['price(in lakhs)','status','builder'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(feartures,target,train_size=0.8,random_state=42)
```

```python
x_train[:5]
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
acc=r2_score(y_pred,y_test)
acc
```

```python
root_mean_squared_error(y_pred,y_test)
```

```python
model = Pipeline([
    ('encoder', OneHotEncoder(handle_unknown='ignore')),
    ('regressor', LinearRegression())
])
```

```python
pk.dump(model,open('house_price_prediction.pkl','wb'))
```
