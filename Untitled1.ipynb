```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from xgboost import XGBRegressor
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Yellow_Taxi_Trip_Data.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['airport_fee'].fillna(0,inplace=True)
df['congestion_surcharge'].fillna(0,inplace=True)
```

```python
df['passenger_count'].mode()
```

```python
df['passenger_count'].fillna(1,inplace=True)
```

```python
df['RatecodeID'].fillna(99,inplace=True)
```

```python
df.columns
```

```python
df=df.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
y=df['tip_amount']
x=df.drop(columns=['tip_amount','tpep_dropoff_datetime','tpep_pickup_datetime','store_and_fwd_flag'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
mask = y_train.notna() & np.isfinite(y_train)

x_train = x_train.loc[mask]
y_train = y_train.loc[mask]
```

```python
model=XGBRegressor(enable_categorical=True)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred[:10]
```

```python
y_test[:10]
```

```python
r2_score(y_pred,y_test)
```
